# Top 15 countries in population

```python
# Function to add value labels to the bars
def add_value_labels(plot):
    for p in plot.patches:
        plot.annotate(format(p.get_width(), '.1f'),
                      (p.get_width(), p.get_y() + p.get_height() / 2.),
                      ha='center', va='center',
                      xytext=(15, 0), textcoords='offset points')

# Define specific colors
custom_colors_1999 = ['#FF5733', '#FFBD33', '#DBFF33', '#75FF33', '#33FF57', 
                      '#33FFBD', '#33DBFF', '#3375FF', '#5733FF', '#BD33FF'] * 3  # Adjust length if necessary
custom_colors_1989 = ['#FFC300', '#FF5733', '#C70039', '#900C3F', '#581845', 
                      '#DAF7A6', '#FFC300', '#FF5733', '#C70039', '#900C3F'] * 3  # Adjust length if necessary

# Assume 'top_15_countries' is derived from the data for top 10 or 15 countries
top_ten_countries = df.nlargest(15, '1999')

# Create the figure
plt.figure(figsize=(15, 14))

# Subplot 1: Top 15 Countries in 1999
plt.subplot(2, 1, 1)
plot = sns.barplot(x='1999', y='Country Name', data=top_ten_countries, palette=custom_colors_1999)
plt.title('Top 15 Countries of Total Population (1999)')
plt.xlabel('Total Population')
plt.ylabel('Country')
add_value_labels(plot)

# Subplot 2: Top 15 Countries in 1989
plt.subplot(2, 1, 2)
plot = sns.barplot(x='1989', y='Country Name', data=top_ten_countries, palette=custom_colors_1989)
plt.title('Top 15 Countries of Total Population in 1989')
plt.xlabel('Total Population')
plt.ylabel('Country')
add_value_labels(plot)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Function to add value labels to the bars
def add_value_labels(plot):
    for p in plot.patches:
        plot.annotate(format(p.get_height(), '.1f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='center',
                      xytext=(0, 10), textcoords='offset points')


# Define specific colors
custom_colors_1999 = ['#FF5733', '#FFBD33', '#DBFF33', '#75FF33', '#33FF57', 
                      '#33FFBD', '#33DBFF', '#3375FF', '#5733FF', '#BD33FF'] * 2  # Adjust length if necessary
custom_colors_1989 = ['#FFC300', '#FF5733', '#C70039', '#900C3F', '#581845', 
                      '#DAF7A6', '#FFC300', '#FF5733', '#C70039', '#900C3F'] * 2  # Adjust length if necessary

# Select the countries with the smallest populations
low_population_countries_1999 = df.nsmallest(15, '1999')
low_population_countries_1989 = df.nsmallest(15, '1989')

# Create the figure
plt.figure(figsize=(15, 13))

# Subplot 1: Low Population Countries in 1999
plt.subplot(2, 1, 1)
plot = sns.barplot(x='Country Name', y='1999', data=low_population_countries_1999, palette=custom_colors_1999)
plt.title('Top 15 Countries with Lowest Total Population (1999)')
plt.xlabel('Country')
plt.ylabel('Total Population')
add_value_labels(plot)
plt.xticks(rotation=45)

# Subplot 2: Low Population Countries in 1989
plt.subplot(2, 1, 2)
plot = sns.barplot(x='Country Name', y='1989', data=low_population_countries_1989, palette=custom_colors_1989)
plt.title('Top 15 Countries with Lowest Total Population in 1989')
plt.xlabel('Country')
plt.ylabel('Total Population')
add_value_labels(plot)
plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

# Top population countries 2000-2023

```python
# Function to add value labels to the bars
def add_value_labels(plot):
    for p in plot.patches:
        plot.annotate(format(p.get_width(), '.1f'),
                      (p.get_width(), p.get_y() + p.get_height() / 2),
                      ha='center', va='center',
                      xytext=(10, 0), textcoords='offset points')

# Define a palette with a color gradient
palette = sns.color_palette("coolwarm", 10)  # 10 colors for 10 countries

# Filter the data for years 2000 to 2023
years = [str(year) for year in range(2000, 2023 + 1)]
population_data = df[['Country Name'] + years]

# Reshape data for easier plotting
population_data_melted = pd.melt(population_data, id_vars=['Country Name'], var_name='Year', value_name='Population')

# Sort data to get the top 10 countries by population for each year
top_population_countries = population_data_melted.groupby('Year').apply(lambda x: x.nlargest(10, 'Population')).reset_index(drop=True)

# Create the figure
plt.figure(figsize=(20, 60))  # Adjust height based on the number of years

# Plot for each year
for i, year in enumerate(years, start=1):
    plt.subplot(12, 2, i)  # 12 rows, 2 columns (since there are 24 years)
    plot = sns.barplot(y='Country Name', x='Population', data=top_population_countries[top_population_countries['Year'] == year], palette=palette)
    plt.title(f'Top 10 Countries by Population in {year}')
    plt.xlabel('Total Population')
    plt.ylabel('Country')
    add_value_labels(plot)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

# Low Population countries

```python
# Function to add value labels to the bars
def add_value_labels(plot):
    for p in plot.patches:
        plot.annotate(format(p.get_width(), '.1f'),
                      (p.get_width(), p.get_y() + p.get_height() / 2),
                      ha='center', va='center',
                      xytext=(10, 0), textcoords='offset points')

# Define a palette with a color gradient
palette = sns.color_palette("coolwarm", 10)  # 10 colors for 10 countries

# Filter the data for years 2000 to 2023
years = [str(year) for year in range(2000, 2023 + 1)]
population_data = df[['Country Name'] + years]

# Reshape data for easier plotting
population_data_melted = pd.melt(population_data, id_vars=['Country Name'], var_name='Year', value_name='Population')

# Sort data to get the bottom 10 countries by population for each year
bottom_population_countries = population_data_melted.groupby('Year').apply(lambda x: x.nsmallest(10, 'Population')).reset_index(drop=True)

# Create the figure
plt.figure(figsize=(20, 60))  # Adjust height based on the number of years

# Plot for each year
for i, year in enumerate(years, start=1):
    plt.subplot(12, 2, i)  # 12 rows, 2 columns (since there are 24 years)
    plot = sns.barplot(y='Country Name', x='Population', data=bottom_population_countries[bottom_population_countries['Year'] == year], palette=palette)
    plt.title(f'Bottom 10 Countries by Population in {year}')
    plt.xlabel('Total Population')
    plt.ylabel('Country')
    add_value_labels(plot)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

# Time Series Analysis

```python
us_population = population_data.set_index('Country Name').loc['United States'].dropna()

# Calculate population growth over time
population_growth = us_population.diff()

# Plotting the population growth as a bar plot
plt.figure(figsize=(12, 6))
population_growth.plot(kind='bar', color='lightgreen')
plt.title('Population Growth in United States (2000-2023)')
plt.xlabel('Year')
plt.ylabel('Population Growth')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Selecting columns from '1960' to '2023' (adjust based on your actual column names)
years_columns = df.columns[df.columns.str.isnumeric()]

# Calculate total population for each country
df['Total Population'] = df[years_columns].sum(axis=1)

# Find the top 5 countries with the highest total population
top_5_high_population = df.nlargest(5, 'Total Population')[['Country Name', 'Total Population']]

# Find the bottom 5 countries with the lowest total population
bottom_5_low_population = df.nsmallest(5, 'Total Population')[['Country Name', 'Total Population']]

# Plotting the data
fig, axes = plt.subplots(2, 1, figsize=(7, 10))

# Top 5 High Population Countries
axes[0].bar(top_5_high_population['Country Name'], top_5_high_population['Total Population'], color='blue')
axes[0].set_title('Top 5 Countries with Highest Population')
axes[0].set_xlabel('Country')
axes[0].set_ylabel('Total Population')

# Bottom 5 Low Population Countries
axes[1].bar(bottom_5_low_population['Country Name'], bottom_5_low_population['Total Population'], color='red')
axes[1].set_title('Top 5 Countries with Lowest Population')
axes[1].set_xlabel('Country')
axes[1].set_ylabel('Total Population')

# Rotate x-axis labels for better visibility
axes[0].tick_params(axis='x', rotation=45)
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
